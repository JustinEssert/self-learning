```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def print_csv(filename):
    df = pd.read_csv(filename)
    print(df.head())

if __name__ == "__main__":
    print_csv('AAPL.csv')
```

```python
def plot_csv(filename):
    df = pd.read_csv(filename)
    df['Close'].plot()
    plt.show()


if __name__ == "__main__":
    plot_csv('AAPL.csv')
```
