## Setup

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

## Preprocessing

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder 
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:,1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:,2])
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Build

```python
import keras as kr
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout
```

```python
# Create a network
cf = Sequential()
```

```python
# Define input & first hidden layer
cf.add(Dense(units=6, activation="elu", input_dim=11, kernel_initializer="uniform"))
```

```python
# Define second hidden layer
cf.add(Dense(units=6, activation="elu", kernel_initializer="uniform"))
```

```python
# Define output layer
cf.add(Dense(units=1, activation="sigmoid", kernel_initializer="uniform"))
```

```python
# Compile ANN
cf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
cf.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
y_pred = cf.predict(X_test)
y_pred = (y_pred >= 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
print('acc:')
print((cm[0,0]+cm[1,1])/(cm[0,0]+cm[0,1]+cm[1,0]+cm[1,1]))
```

## Build - K-fold 

```python
import keras as kr

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier

from sklearn.model_selection import cross_val_score
```

```python
def build_classifier():
    # Create a network
    cf = Sequential()
    # Define input & first hidden layer
    cf.add(Dense(units=6, activation="elu", input_dim=11, kernel_initializer="uniform"))
    cf.add(Dropout(.1))
    # Define second hidden layer
    cf.add(Dense(units=6, activation="elu", input_dim=11, kernel_initializer="uniform"))
    cf.add(Dropout(.1))
    # Define output layer
    cf.add(Dense(units=1, activation="sigmoid", kernel_initializer="uniform"))
    # Compile ANN
    cf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return cf
cf = KerasClassifier(build_fn=build_classifier, batch_size=10, epochs=100)
```

```python
# Use 10 folds 
accs = cross_val_score(estimator=cf, X=X_train, y=y_train, cv=10, n_jobs=-1)
```

```python
mean=accs.mean()
var=accs.std()
print(accs)
print('mean', mean)
print('var', var)
```

## Grid Search

```python
import keras as kr

from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.wrappers.scikit_learn import KerasClassifier

from sklearn.model_selection import GridSearchCV
```

```python
def build_classifier(hidden_units):
    # Create a network
    cf = Sequential()
    # Define input & first hidden layer
    cf.add(Dense(units=hidden_units, activation="elu", input_dim=11, kernel_initializer="uniform"))
    cf.add(Dropout(.1))
    # Define second hidden layer
    cf.add(Dense(units=hidden_units, activation="elu", input_dim=11, kernel_initializer="uniform"))
    cf.add(Dropout(.1))
    # Define output layer
    cf.add(Dense(units=1, activation="sigmoid", kernel_initializer="uniform"))
    # Compile ANN
    cf.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return cf
cf = KerasClassifier(build_fn=build_classifier, batch_size=10, epochs=100)
```

```python
params = {'batch_size': [25, 32],
          'epochs': [100, 500],
          'hidden_units': [6, 8, 12]}
```

```python
grid_search = GridSearchCV(estimator=cf, 
                           param_grid=params,
                           scoring = 'accuracy',
                           cv = 10)
grid_search = grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
best_acc = grid_search.best_score_
```
